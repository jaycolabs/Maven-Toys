# Remove Unwanted Columns, First Make Copy

```python
toys_data2 = toys_data.copy()
```

```python
toys_data2
```

```python
toys_data2.columns
```

```python
toys_data2.drop(['Unnamed: 0', 'Sale_ID', 'Store_ID', 'Product_ID','Store_Open_Date'], axis=1, inplace=True)
```

```python
toys_data2
```

# Reorder Dataframe

```python
toys_data2 = toys_data2[['Date','Store_Name', 'Store_City', 'Store_Location', 'Product_Name', 'Product_Category', 'Product_Price', 'Product_Cost', 'Units']]
```

```python
toys_data2
```

# Plotting

```python
# What City Got The Most Transaction?

px.histogram(toys_data2, x="Store_City").update_xaxes(categoryorder="total descending")
```

![Graph1](plotly/1.png)

```python
# What Day Had The Most Transcations?

px.histogram(toys_data2, x="Date")
```

![Graph2](plotly/2.png)

```python
# What Category Did People Like The Most?

px.histogram(toys_data2, x="Product_Category").update_xaxes(categoryorder="total descending")
```

![Graph3](plotly/3.png)

```python
# What City Spent The Most Money?

px.pie(toys_data2, values='Product_Price', names='Store_City')
```

![Graph4](plotly/4.png)

```python
# Which Store Location Sold The Most Units? What Category Is Most Popular?

px.sunburst(toys_data2, path=['Store_Location', 'Product_Category', 'Product_Price'], values='Units')
```
