# Import Packages

```python
import pandas as pd 
import numpy as np
```

# Merging Dataframes

```python
sales = pd.read_csv('sales.csv')
sales
```

```python
products = pd.read_csv('products.csv')
products
```

```python
sales_products = pd.merge(sales, products, on='Product_ID')
```

```python
sales_products
```

```python
sales_products2 = sales_products.copy()
```

```python
sales_products2
```

```python
stores = pd.read_csv('stores.csv')
stores
```

```python
toys = pd.merge(sales_products2, stores, on='Store_ID')
```

```python
toys
```

# Export 

```python
toys.to_csv('toys_data.csv')
```
